```python
#events class
def to_event(timestamp,device,phase):
    timestamp=datetime.strptime(timestamp[:23], '%m/%d/%Y %H:%M:%S.%f')
    t=new_start_time 
    event_time=(timestamp - t).total_seconds()
    return [device,event_time,phase,0]
#     def __str__(self):
#         return 'Dno=%3s datetime=%12s P=%c F=%s'%(self.device,self.event_time,self.phase,str(self.files))
```

```python
def read_events():
    events=pd.DataFrame(columns=['device','time','phase','file'])
    #opening the csv file and referring  it with the variable f
    with open("event_list_2.txt","r") as datafile:
        #skipping the first line that contains the headers
        datafile.readline()
        #reading the contents of the file and serepating each line with a delimmiter 'COMMA'
        lines = csv.reader(datafile,delimiter=',')
        #printing words one by one from each line of the csvfile lines
        i=0
        for line in lines:
                events.loc[i]=to_event(line[0],line[1],line[2])
                i+=1
    return events
```
