```python
import os 
import numpy as np
import pandas as pd
```

```python
files=['events/derived/'+f+'/'+ff for f in os.listdir('events/derived') if 'dataset' in f for ff in os.listdir('events/derived/'+f) ]
```

```python
files
```

```python
df=pd.DataFrame(columns=['p','q','s','device'])
c=0
for f in files:
    try:
        data=np.loadtxt(f,delimiter=',')[:,2:]
    #     print(data)
        data=list(data[-2]-data[1])
        data.append(f.split('/')[3].split('_')[0])
    #     input(data)
        df.loc[c]=data
        print(c)
        c+=1
    except:
        print(data)
        input()
```

```python
count=df.groupby('device').count()
```

```python
single_events=count[count.p==1].index
```

```python
for se in single_events:
    df=df[df.device!=se]
```

```python
df.to_csv('events/PQS.csv',sep='\t')
```
