```python
import pandas as pd
import os
```

```python
data_list=['../'+fol+'/'+f for fol in os.listdir('../') for f in os.listdir('../'+fol) if 'ivdata' in f and '.bz2' not in f and '.swp' not in f]
data_sets=[int(f.split('_')[-1].split('.')[0]) for f in data_list]
print(data_list)
print(data_sets)
```

```python
def find_se_time():
    se_time=pd.DataFrame(columns=['start','end'])
    for i in range(len(data_sets)):
        print(data_list[i])
        with open (data_list[i],'rb') as f:
            #skipping the first 24 lines (headers)
            for x in range(24):
                f.readline()
            #saving the position and reading the first line
            pos,a = f.tell(),float(f.readline().decode("utf-8").split(',')[0])
            #Readoing the last line of the file
            c = f.readline().decode("utf-8")
            f.seek(-80,2)
            b=''
            while True:
                c=f.readline()
                if len(c)==0:
                    break
                b=c
            x=b.decode("utf-8")
            b=float(x.split(',')[0])
            
            se_time.loc[data_sets[i]]=[a,b]
    return se_time
```

```python
se_time=find_se_time()
```

```python
se_time.sort_index().to_csv('start_end.csv',sep='\t')
```
