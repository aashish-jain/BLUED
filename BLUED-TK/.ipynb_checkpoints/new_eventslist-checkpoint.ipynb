```python
start_datetime = datetime.strptime('2011/10/20 11:58:32.623', '%Y/%m/%d %H:%M:%S.%f')

offset=blued.offset


file_list={}
for i in range(1,13):
    file_list[i]=((i-1)*400+1,i*400)
file_list[13]=(4801,5230)
file_list[14]=(5231,5630)
file_list[15]=(5631,6030)
file_list[16]=(6031,6430)


#frequency
freq = 12000
#settling the time before and after events
time_before_event=0.5
time_after_event=1
```

```python
#events class
def to_event(timestamp,device,phase):
        event_time=(datetime.strptime(timestamp[:23], '%m/%d/%Y %H:%M:%S.%f') - start_datetime).total_seconds()
        return [device,event_time,phase,0]
#     def __str__(self):
#         return 'Dno=%3s datetime=%12s P=%c F=%s'%(self.device,self.event_time,self.phase,str(self.files))
```

```python
def read_events():
    events=pd.DataFrame(columns=['device','time','phase','file'])
    #opening the csv file and referring  it with the variable f
    with open("events_list.txt","r") as datafile:
        #skipping the first line that contains the headers
        datafile.readline()
        #reading the contents of the file and serepating each line with a delimmiter 'COMMA'
        lines = csv.reader(datafile,delimiter=',')
        #printing words one by one from each line of the csvfile lines
        i=0
        for line in lines:
                events.loc[i]=to_event(line[0],line[1],line[2])
                i+=1
    return events
```

```python
events=read_events()
```
